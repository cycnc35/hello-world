```python
import sounddevice as sd
import scipy.io.wavfile
import matplotlib.pyplot as plt
%matplotlib inline
duration = 4 # seconds
CHANNELS = 1 # audio channels
fs = 16000 # sampling rate
s = sd.rec(int(duration * fs), samplerate=fs,channels=CHANNELS); sd.wait()
N=1024 # fft length, power of 2 is good
L=800 # window size, so zero padding = N-L.
ovl = L-L/8 # overlap of successive windows
Pxx,freqs,bins,im=plt.specgram(s[:,0], NFFT=L, Fs=fs,pad_to=N, noverlap=ovl)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

plt.figure(figsize=(10,6))
ys = 200 + np.random.randn(1000)
x = [x for x in range(len(ys))]
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.plot(x, ys, '-')
plt.show()
```
